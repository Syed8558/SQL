```python
1.Write a SQL query to display employee first and last names in uppercase.
```

```python
SELECT
    UPPER(first_name) AS FIRST_NAME,
    UPPER(last_name)  AS LAST_NAME
FROM employees;
```

```python
2.Write a query to extract the year from the `hire_date` of each employee.
```

```python
SELECT
    emp_no,
    EXTRACT(YEAR FROM hire_date) AS hire_year
FROM employees;
```

```python
3. Write a query to display employee full name by concatenating `first_name` and `last_name`.
```

```python
SELECT
    first_name  as FIRST_NAME
    last_name as LAST_NAME
    FIRST_NAME||' '||LAST_NAME as FULL_NAME
FROM employees;
```

```python
4.Write a query to find the length of each employeeâ€™s first name.
```

```python
SELECT
    first_name || ' ' || last_name AS full_name,
    LENGTH(first_name || ' ' || last_name) AS full_name_length
FROM employees;
```

```python
5. Write a query to replace the letter 'a' with 'A' in employee last names.
```

```python
SELECT
    last_name,
    REPLACE(last_name, 'a', 'A') AS updated_last_name
FROM employees;
```

```python
6. Write a query to display employee age using `birth_date`.
```

```python
SELECT
    emp_no,
    first_name,
    last_name,
    TRUNC(MONTHS_BETWEEN(SYSDATE, birth_date) / 12) AS age
FROM employees;
```

```python
7. Write a query to round employee salaries to the nearest thousand.
```

```python
SELECT
    emp_no,
    salary,
    ROUND(salary, -3) AS rounded_salary
FROM salaries;
```

```python
8. Write a query to display the month name of employee hire dates.
```

```python
SELECT
    emp_no,
    hire_date,
    TO_CHAR(hire_date, 'Month') AS hire_month
FROM employees;
```

```python
9. Write a query to handle NULL `to_date` values in salary records using `COALESCE`.
```

```python
SELECT
    emp_no,
    salary,
    from_date,
    COALESCE(to_date, CURRENT_DATE) AS to_date_handled
FROM salaries;
```

```python
10. Write a query to return NULL when `from_date` and `to_date` are the same using `NULLIF`.
```

```python
SELECT
    emp_no,
    salary,
    from_date,
    to_date,
    NULLIF(from_date, to_date) AS date_check
FROM salaries;
```

```python
11.Write a SQL query to find employees earning more than the average salary.
```

```python
SELECT
    e.emp_no,
    e.first_name,
    e.last_name,
    s.salary
FROM employees e
JOIN salaries s
  ON e.emp_no = s.emp_no
WHERE s.salary > (
    SELECT AVG(salary)
    FROM salaries
);
```

```python
12.Write a query to list employees who are currently working in the 'Sales' department.
```

```python
SELECT
    e.emp_no,
    e.first_name,
    e.last_name
FROM employees e
JOIN dept_emp de
    ON e.emp_no = de.emp_no
JOIN departments d
    ON de.dept_no = d.dept_no
WHERE d.dept_name = 'Sales'
  AND de.to_date IS NULL;
```

```python
13.Write a query to find employees who have never been managers.
```

```python
SELECT
    e.emp_no,
    e.first_name,
    e.last_name
FROM employees e
WHERE NOT EXISTS (
    SELECT 1
    FROM dept_manager dm
    WHERE dm.emp_no = e.emp_no
);
```

```python
14.Write a query to display employees whose salary is greater than their departmentâ€™s average salary.
```

```python
SELECT e.emp_no, e.first_name, e.last_name
FROM employees e
WHERE e.emp_no IN (
    SELECT s.emp_no
    FROM salaries s
    JOIN dept_emp de ON s.emp_no = de.emp_no
    WHERE s.to_date = '9999-01-01'
      AND de.to_date = '9999-01-01'
      AND s.salary > (
          SELECT AVG(s2.salary)
          FROM salaries s2
          JOIN dept_emp de2 ON s2.emp_no = de2.emp_no
          WHERE de2.dept_no = de.dept_no
            AND s2.to_date = '9999-01-01'
            AND de2.to_date = '9999-01-01'
      )
);
```

```python
15.Write a query to find the employee(s) with the maximum salary.
```

```python
SELECT e.emp_no, e.first_name, e.last_name
FROM employees e
WHERE e.emp_no IN (
    SELECT emp_no
    FROM salaries
    WHERE salary = (SELECT MAX(salary) FROM salaries)
);
```

```python
20. Write a query to display employees hired earlier than their managers.
```

```python
SELECT e.emp_no, e.first_name, e.last_name
FROM employees e
JOIN dept_emp de ON e.emp_no = de.emp_no
JOIN dept_manager dm ON de.dept_no = dm.dept_no
JOIN employees m ON dm.emp_no = m.emp_no
WHERE de.to_date = '9999-01-01'
  AND dm.to_date = '9999-01-01'
  AND e.hire_date < m.hire_date;
```

```python
21. Write a query to find the current manager of each department.
```

```python
SELECT d.dept_no,
       d.dept_name,
       e.emp_no,
       e.first_name,
       e.last_name
FROM departments d
JOIN dept_manager dm ON d.dept_no = dm.dept_no
JOIN employees e ON dm.emp_no = e.emp_no
WHERE dm.to_date = '9999-01-01';
```

```python
22. Write a query to display the current salary of each employee.
```

```python
SELECT e.emp_no,
       e.first_name,
       e.last_name,
       s.salary
FROM employees e
JOIN salaries s ON e.emp_no = s.emp_no
WHERE s.to_date = '9999-01-01';
```

```python
23. Write a query to find the top 5 highest-paid employees.
```

```python
SELECT e.emp_no,
       e.first_name,
       e.last_name,
       s.salary
FROM employees e
JOIN salaries s ON e.emp_no = s.emp_no
WHERE s.to_date = '9999-01-01'
ORDER BY s.salary DESC
FETCH FIRST 5 ROWS ONLY;
```

```python
24. Write a query to calculate average salary per department.
```

```python
SELECT d.dept_name,
       AVG(s.salary) AS avg_salary
FROM departments d
JOIN dept_emp de ON d.dept_no = de.dept_no
JOIN salaries s ON de.emp_no = s.emp_no
WHERE de.to_date = '9999-01-01'
  AND s.to_date = '9999-01-01'
GROUP BY d.dept_name;
```

```python
30. Write a query to list managers who manage more than 5 employees.
```

```python
SELECT dm.emp_no,
       COUNT(de.emp_no) AS employee_count
FROM dept_manager dm
JOIN dept_emp de ON dm.dept_no = de.dept_no
WHERE dm.to_date = '9999-01-01'
  AND de.to_date = '9999-01-01'
GROUP BY dm.emp_no
HAVING COUNT(de.emp_no) > 5;
```
